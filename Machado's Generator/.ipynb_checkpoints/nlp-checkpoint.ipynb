```python
from nltk.corpus import machado
text = machado.words('romance/marm05.txt')

print "Número de palavras: ",len(text)
print "Número de palavras diferentes: ",len(set(text))
```

```python
from nltk import ngrams, FreqDist
target = 'olhos'
fd = FreqDist(ng for ng in ngrams(text,5) if target in ng)

for hit in fd.elements():
    print(' '.join(hit))
```
